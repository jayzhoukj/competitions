```
os.environ['TZ'] = 'Singapore'
time.tzset()

df['1st_deliver_attempt'] = df['1st_deliver_attempt'].apply(lambda x: time.strftime('%Y-%m-%d', time.localtime(x)))
df
```

```
import math

def convert_time(data):
    if math.isnan(data):
        return np.NaN
    else:
        return time.strftime('%Y-%m-%d', time.localtime(data))

df['2nd_deliver_attempt'] = df['2nd_deliver_attempt'].apply(convert_time)
df
```

```
def extract_area(data):
    add_info = data.lower().split(" ")

    if add_info[-1] == 'manila':
        return add_info[-2].title() + " " + add_info[-1].title()
    else:
        return add_info[-1].title()

df['buyeraddress'] = df['buyeraddress'].apply(extract_area)
df
```

```
df['selleraddress'] = df['selleraddress'].apply(extract_area)
df
```

```
# YYYY-MM-DD
holiday_list = ['2020-03-08', '2020-03-25', '2020-03-30', '2020-03-31']

def first_attempt_time(row):
    return np.busday_count(row['pick'], row['1st_deliver_attempt'], weekmask = '1111110', holidays = holiday_list)

def second_attempt_time(row):
    try:
        return np.busday_count(row['1st_deliver_attempt'], row['2nd_deliver_attempt'], weekmask = '1111110', holidays = holiday_list)
    except:
        return np.NaN
```

```
df['first_time_taken'] = df.apply(first_attempt_time, axis=1)
df['second_time_taken'] = df.apply(second_attempt_time, axis=1)
df
```

```
def get_allowance(row):
    return int(df_sla.loc[row['selleraddress'], row['buyeraddress']].split(" ")[0])

df['sla_allowance'] = df.apply(get_allowance, axis=1)
df
```

```
def check_late(row):
    if row['second_time_taken'] > 3:
        return 1
    else:
        if row['first_time_taken'] <= row['sla_allowance']:
            return 0
        else:
            return 1

df['is_late'] = df.apply(check_late, axis=1)
df
```

```
df['is_late'].value_counts()
```

```
df_submission = df.loc[:, ['orderid', 'is_late']]
df_submission
```
