## Transform Data to Required Formats and Information

```python
transformed = base.copy()
```

```python
transformed.head()
```

```python
def pick(row):
    try:
        return time.strftime('%Y-%m-%d', time.localtime(row['pick']))
    except:
        return
    
def first(row):
    try:
        return time.strftime('%Y-%m-%d', time.localtime(row['1st_deliver_attempt']))
    except:
        return
    
def second(row):
    try:
        return time.strftime('%Y-%m-%d', time.localtime(row['2nd_deliver_attempt']))
    except:
        return
```

```python
# Change dates from epoch to string
transformed['pick'] = transformed.apply(pick, axis=1)
transformed['1st_deliver_attempt'] = transformed.apply(first, axis=1)
transformed['2nd_deliver_attempt'] = transformed.apply(second, axis=1)
```

```python
transformed.head()
```

```python
transformed['buyeraddress'] = transformed['buyeraddress'].str.upper()
transformed['selleraddress'] = transformed['selleraddress'].str.upper()
```

```python
transformed.head()
```

```python
def extract_area(data):
    add_info = data.split(" ")

    if add_info[-1] == 'MANILA':
        return add_info[-2].title() + " " + add_info[-1].title()
    else:
        return add_info[-1].title()
```

```python
test = transformed['selleraddress'].copy()
transformed['selleraddress'] = test.apply(extract_area)
test = transformed['buyeraddress'].copy()
transformed['buyeraddress'] = test.apply(extract_area)
```

```python
transformed.head()
```

## Calculate 3PL Time Taken

```python
final = transformed.copy()
```

```python
final
```

```python
# YYYY-MM-DD
holiday_list = ['2020-03-08', '2020-03-25', '2020-03-30', '2020-03-31']
```

```python
def first_attempt_time(row):
    return np.busday_count(row['pick'], row['1st_deliver_attempt'], weekmask = '1111110', holidays = holiday_list)

def second_attempt_time(row):
    try:
        return np.busday_count(row['1st_deliver_attempt'], row['2nd_deliver_attempt'], weekmask = '1111110', holidays = holiday_list)
    except:
        return
```

```python
final['first_time_taken'] = final.apply(first_attempt_time, axis=1)
```

```python
final.head(10)
```

```python
final['second_time_taken'] = final.apply(second_attempt_time, axis=1)
```

```python
final.head(10)
```

```python
def first_check(row):
    if row['selleraddress'] == 'Visayas' or row['selleraddress'] == 'Mindanao':
        alloted_time = 7
    
    elif row['buyeraddress'] == 'Visayas' or row['buyeraddress'] == 'Mindanao':
        alloted_time = 7
    
    elif row['selleraddress'] == 'Metro Manila':
        # Metro Manila (3), Luzon (5), Visayas (7), Mindanao (7)
        if row['buyeraddress'] == 'Metro Manila':
            alloted_time = 3
        else:
            alloted_time = 5            
    else:
        alloted_time = 5
    
    if row['first_time_taken'] > alloted_time:
        return 1
    else:
        return 0
```

```python
def second_check(row):
    try:
        if row['second_time_taken'] > 3:
            return 1
        else:
            return 0
    except:
        return 0
```

```python
final['first_check'] = final.apply(first_check, axis = 1)
```

```python
final.head(10)
```

```python
final['second_check'] = final.apply(second_check, axis = 1)
```

```python
final.head(50)
```

```python
final['is_late'] = ((final['first_check'] + final['second_check']) >= 1).astype(int)
```

```python
final.head(10)
```
